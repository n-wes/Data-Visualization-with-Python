```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify as sq
import seaborn as sns

data = pd.read_csv('top50.csv', encoding = 'cp1252')
```

```python
# Our Dataset
data = data.rename(columns={'Unnamed: 0':'Rank'})
data.head(50)
```

```python
data.describe()
```

```python
# Show top artists
artist_frequency = data['Artist.Name'].value_counts()
artist_frequency.sort_values().plot(kind='barh',xticks=[n for n in range(0, max(artist_frequency)+1)],color='#7FDBFF',edgecolor='black',figsize=(10,10))
plt.ylabel('Artists')
plt.xlabel('Occurences in Top 50')
plt.title('Occurences of Artists in Top 50')
plt.show()
```

```python
# Show top genres
genre_frequency = data['Genre'].value_counts()
genre_frequency.sort_values().plot(kind='barh',xticks=[n for n in range(0, max(genre_frequency)+1)],color='#FFDC00',edgecolor='black',figsize=(10,10))
plt.ylabel('Genre')
plt.xlabel('Occurences in Top 50')
plt.title('Occurences of Genres in Top 50')
plt.show()
```

```python
# Show genre frequency using square plot
plt.figure(figsize=(14,8))
genre_counts = data['Genre'].value_counts()
sq.plot(sizes=genre_counts.array,label=genre_counts.index,alpha=.6)
plt.axis('off')
plt.show()
```

```python
# Show artist frequency using square plot
plt.figure(figsize=(14,8))
artist_counts = data['Artist.Name'].value_counts()
sq.plot(sizes=artist_counts.array,label=artist_counts.index,alpha=.6)
plt.axis('off')
plt.show()
```

```python
# Show Correlations
sns.pairplot(data)
plt.show()
```
