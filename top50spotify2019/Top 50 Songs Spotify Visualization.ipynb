```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify as sq
import seaborn as sns

data = pd.read_csv('top50.csv', encoding = 'cp1252')
data = data.rename(columns={'Unnamed: 0':'Rank'})

def htmap_trunc_fix():
    ### Fix heatmap truncation problem
    b, t = plt.ylim() # discover the values for bottom and top
    b += 0.5 # Add 0.5 to the bottom
    t -= 0.5 # Subtract 0.5 from the top
    plt.ylim(b, t) # update the ylim(bottom, top) values
    ###
```

```python
# Our Dataset
data.head(50)
```

```python
data.describe()
```

```python
# Show top artists
artist_frequency = data['Artist.Name'].value_counts()
artist_frequency.sort_values().plot(kind='barh',xticks=[n for n in range(0, max(artist_frequency)+1)],color='#7FDBFF',edgecolor='black',figsize=(10,10))
plt.ylabel('Artists')
plt.xlabel('Occurences in Top 50')
plt.title('Occurences of Artists in Top 50')
plt.show()
```

```python
# Show top genres
genre_frequency = data['Genre'].value_counts()
genre_frequency.sort_values().plot(kind='barh',xticks=[n for n in range(0, max(genre_frequency)+1)],color='#FFDC00',edgecolor='black',figsize=(10,10))
plt.ylabel('Genre')
plt.xlabel('Occurences in Top 50')
plt.title('Occurences of Genres in Top 50')
plt.show()
```

```python
# Show genre frequency using square plot
plt.figure(figsize=(14,8))
genre_counts = data['Genre'].value_counts()
sq.plot(sizes=genre_counts.array,label=genre_counts.index,alpha=.6)
plt.axis('off')
plt.show()
```

```python
# Show artist frequency using square plot
plt.figure(figsize=(14,8))
artist_counts = data['Artist.Name'].value_counts()
sq.plot(sizes=artist_counts.array,label=artist_counts.index,alpha=.6)
plt.axis('off')
plt.show()
```

```python
# Pairplots
sns.pairplot(data)
plt.show()
```

```python
# Pearson Correlation
# Linear relationship correlation
p_corr = data.corr(method='pearson') 
p_corr
```

```python
plt.figure(figsize=(15,10))
plt.title('Pearson Correlation Heatmap')
sns.heatmap(p_corr,annot=True,vmin=-1,vmax=1,cmap="rocket",center=1)
htmap_trunc_fix()
plt.show()
```

```python
# Spearman Correlation
# Monotonic relationship correlation
sp_corr = data.corr(method='spearman') 
sp_corr
```

```python
plt.figure(figsize=(15,10))
plt.title('Spearman Correlation Heatmap')
sns.heatmap(sp_corr,annot=True,vmin=-1,cmap="twilight_shifted_r",vmax=1,center=1)
htmap_trunc_fix()
plt.show()
```
