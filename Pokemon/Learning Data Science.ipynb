```python
pokemon.Speed.plot(kind = 'line', color = 'b', label = 'Speed', linewidth = 1, alpha = 0.5, grid = True, linestyle = ':')
pokemon.Defense.plot(color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc = 'upper right')
plt.xlabel('Pokemon #')
plt.ylabel('Stat Strength')
plt.title('Speed and Defense')
plt.show()
```

```python
pokemon.plot(kind = 'scatter', x = 'Attack', y = 'Defense', alpha = 0.5, color = 'b')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack Defense Scatter Plot')
plt.show()
```
