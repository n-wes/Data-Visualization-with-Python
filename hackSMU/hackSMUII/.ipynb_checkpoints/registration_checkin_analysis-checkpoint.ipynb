```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
check_ins = pd.read_excel('check_ins.xlsx')
registrations = pd.read_excel('registrations.xlsx')
submissions = pd.read_csv('submissions.csv')

check_ins.loc[check_ins['Registered'] == "Yes, I signed up ahead of time", "Registered"] = "Yes"
check_ins.loc[check_ins['Registered'] == "No, I'm a walk-on", "Registered"] = "No"
```

```python
registrations['School'].value_counts().rename_axis('School').reset_index(name='Count')
```

```python
fig, ax = plt.subplots(figsize=(20, 15))
registrations['School'].value_counts(ascending=True).plot.barh(title='Registrants by School')
ax.set_xlabel('Count')

total = len(registrations['School'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))
for spine in ax.spines:
    ax.spines[spine].set_visible(False)
        
plt.savefig('reg_school_count.png', bbox_inches = 'tight')
```

```python
check_ins['University'].value_counts().rename_axis('School').reset_index(name='Count')
```

```python
fig, ax = plt.subplots(figsize=(20,15))
check_ins['University'].value_counts(ascending=True).plot.barh(title='Participants by School')
ax.set_xlabel('Count')

total = len(check_ins['University'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

for spine in ax.spines:
    ax.spines[spine].set_visible(False)

plt.savefig('checkin_count.png', bbox_inches = 'tight')
```

```python
fig, ax = plt.subplots(figsize=(10,7))
registrations['School Type'] = registrations.apply(lambda row: "High School" if any(x in row['School'] for x in ["High School", "TAMS", "Academy"]) else "College" , axis=1)
registrations['School Type'].value_counts().plot.pie(title='College vs High School Registrants', colors=['#6c5b7b', '#f8b195'], autopct='%1.0f%%')
ax.set_ylabel('')
plt.savefig('registrations_school_pie.png')
```

```python
registrations['School Type'].value_counts().rename_axis('School Type').reset_index(name='Count').transpose()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
check_ins['School Type'] = check_ins.apply(lambda row: "High School" if any(x in row['University'] for x in ["High School", "TAMS", "Academy"]) else "College" , axis=1)
check_ins['School Type'].value_counts().plot.pie(title='College vs High School Participants', colors=['#99b898','#fecea8'], autopct='%1.0f%%')
ax.set_ylabel('')
plt.savefig('checkins_school_pie.png')
```

```python
check_ins['School Type'].value_counts().rename_axis('School Type').reset_index(name='Count').transpose()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
check_ins['Pre-Registered'] = check_ins.apply(lambda row: "Registered" if row["Registered"] == "Yes" else "Walk-in", axis=1)
check_ins['Pre-Registered'].value_counts().plot.pie(title="Registered vs Walk-In Participants", colors=['#355c7d','#f8b195'], autopct='%1.0f%%')
ax.set_ylabel('')
plt.savefig('reg_vs_walk.png')
```
